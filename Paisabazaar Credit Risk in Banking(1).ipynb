# Project Summary – Paisabazaar Credit Score Analysis 

This project analyzes customer credit behavior using the **Paisabazaar Credit Risk dataset** with the objective of understanding the factors influencing **Credit Score classification (Good, Standard, Poor)**. The analysis is entirely driven by **data visualizations**, enabling clear interpretation of customer financial patterns and risk levels.

## Univariate Analysis

Univariate analysis focuses on understanding the distribution and behavior of individual variables in the dataset. The age variable (`age_in_years`) helps identify the dominant customer segment. Most customers fall within economically active age groups, indicating higher engagement with credit products and stable repayment potential. This insight supports the design of age-specific financial offerings.

The distribution of credit amount (`credit_amount`) reveals customer borrowing patterns. The majority of customers take moderate loan amounts, while a smaller group opts for higher credit values, indicating increased risk exposure. This helps in assessing affordability and identifying over-leveraged customers.

Loan duration (`duration_in_months`) shows customer preferences for repayment periods. Short and medium-term durations are more common, suggesting manageable repayment behavior, whereas longer durations may reflect higher credit dependency.

Other individual variables such as installment rate, number of existing credits, and length of residence provide additional insights into financial stability. Extreme values in these variables can indicate higher credit risk.

## Bivariate Analysis

Bivariate analysis examines relationships between pairs of variables to uncover credit risk patterns. The relationship between age and credit amount suggests that middle-aged customers generally borrow higher amounts, supported by greater financial stability.

Analysis of credit amount and loan duration shows that larger loans are typically associated with longer repayment periods, increasing long-term exposure for the lender.

The interaction between the number of existing credits and credit amount indicates that customers with multiple active credits often face higher financial burden, increasing default risk. Higher installment rates further signal repayment stress.

Employment duration and housing status also demonstrate meaningful relationships with credit behavior, where longer employment history and stable housing conditions are linked to lower credit risk.

## Multivariate Analysis

Multivariate analysis provides a comprehensive view of credit risk by evaluating multiple variables simultaneously. Creditworthiness is influenced by the combined effect of age, employment stability, credit amount, loan duration, and existing financial obligations.

Customers with stable employment, fewer existing credits, moderate installment rates, and reasonable loan durations typically exhibit lower risk profiles. Conversely, high loan amounts combined with multiple credits and short employment history increase the likelihood of default.

Correlation patterns among numerical variables highlight strong relationships between credit amount, loan duration, and installment rates, validating their importance in credit risk modeling and predictive analysis.

### Data Wrangling

```python
# 1) Remove hidden spaces in column names
bank_df.columns = bank_df.columns.str.strip()


# 2) Drop ID column (only useless column here)
bank_df.drop(columns=['id'], inplace=True)


# 3) Rename columns for easy handling
bank_df.rename(columns={
    'Duration in months': 'Duration',
    'Credit amount': 'Credit_amount',
    'Age in years': 'Age',
    'Present employment (years)': 'Employment_years',
    'Installment rate in % of disposable income': 'Installment_rate',
    'Present residence since X years': 'Residence_years'
}, inplace=True)


# 4) Remove outliers from Credit_amount using IQR
Q1 = bank_df['Credit_amount'].quantile(0.25)
Q3 = bank_df['Credit_amount'].quantile(0.75)
IQR = Q3 - Q1

bank_df = bank_df[
    (bank_df['Credit_amount'] >= (Q1 - 1.5 * IQR)) &
    (bank_df['Credit_amount'] <= (Q3 + 1.5 * IQR))
]


# 5) Check & fill missing values (if any)
bank_df.fillna(method='ffill', inplace=True)


# 6) Convert datatypes
bank_df['Age'] = bank_df['Age'].astype(int)
bank_df['Credit_amount'] = bank_df['Credit_amount'].astype(int)
bank_df['Duration'] = bank_df['Duration'].astype(int)


# 7) Create new feature: Credit per Month
bank_df['Credit_per_month'] = bank_df['Credit_amount'] / bank_df['Duration']


# 8) Create Age Group feature
bank_df['Age_group'] = pd.cut(
    bank_df['Age'],
    bins=[18, 30, 45, 60, 100],
    labels=['Young', 'Adult', 'Mature', 'Senior']
)


# 9) Convert target variable to numeric
bank_df['Status'] = bank_df['Status'].map({'Good': 0, 'Bad': 1})


# 10) Final check
bank_df.info()
bank_df.head()
```

## **Univariate Analysis**

#### Loan Status Count Distribution

```python
# Chart - 1 visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Use correct column
status_col = 'Status of savings account/bonds'

# Strip any leading/trailing spaces from column names
bank_df.columns = bank_df.columns.str.strip()

if status_col not in bank_df.columns:
    raise ValueError(f"Column '{status_col}' not found! Available columns:\n{bank_df.columns}")

# Drop rows with NaN in the chosen column
plot_df = bank_df.dropna(subset=[status_col])

# Countplot
plt.figure(figsize=(11,5))

order = plot_df[status_col].value_counts().index
ax = sns.countplot(x=status_col, data=plot_df, order=order)

plt.title("Loan Status Count Distribution", fontsize=14, weight='bold')
plt.xlabel("Status", fontsize=12)
plt.ylabel("Count", fontsize=12)

#  Color bars
labels = [t.get_text() for t in ax.get_xticklabels()]

# Define custom colors
color_map = {
    'good': '#1f77b4',  # Blue
    'bad': '#ff7f0e'    # Orange
}

for bar, label in zip(ax.patches, labels):
    height = bar.get_height()
    x = bar.get_x() + bar.get_width()/2

    # Assign color based on category
    if "good" in label.lower():
        bar.set_color(color_map['good'])
        text = "Good\nLow Risk"
    else:
        bar.set_color(color_map['bad'])
        text = "Bad\nHigh Risk"

    ax.text(
        x, height*0.9,
        text,
        ha='center',
        va='top',
        color='white',
        fontsize=10,
        weight='bold'
    )

# Insight text
status_counts = plot_df[status_col].value_counts()
total = status_counts.sum()

good_percent = round((status_counts[status_counts.index.str.lower().str.contains('good')].sum()/total)*100, 1)
bad_percent = round((status_counts[status_counts.index.str.lower().str.contains('bad')].sum()/total)*100, 1)

insight_text = (
    "Insight:\n\n"
    f"• {good_percent}% customers are in Good category → Low default risk\n\n"
    f"• {bad_percent}% customers are in Bad category → High default risk\n\n"
    "• Majority of borrowers are reliable.\n"
    "• Useful for credit risk profiling and loan approval strategy."
)

plt.subplots_adjust(right=0.68)

plt.text(1.02, 0.5, insight_text,
         transform=ax.transAxes,
         fontsize=11,
         va='center')

plt.show()

```

## **Multivariate Analysis**

#### Correlation Heatmap of Numeric Features

```python
# Chart - 8 visualization 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

# Prepare data
numeric_df = bank_df.select_dtypes(include='number')
corr = numeric_df.corr()

# Find strongest correlation 
corr_matrix = corr.copy()
np.fill_diagonal(corr_matrix.values, 0)
max_corr = corr_matrix.abs().unstack().idxmax()
value = corr_matrix.loc[max_corr]

# Figure with two sections 
fig = plt.figure(figsize=(14,8))
gs = GridSpec(1, 2, width_ratios=[3, 1])

# Heatmap axis
ax1 = fig.add_subplot(gs[0])
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=ax1)
ax1.set_title("Correlation Heatmap of Numeric Features")

# Insight axis 
ax2 = fig.add_subplot(gs[1])
ax2.axis('off')

insight_text = (
    "Insight:\n\n"
    f"Strongest correlation:\n"
    f"{max_corr[0]} ↔ {max_corr[1]} ({value:.2f})\n\n"
    "Highly correlated features indicate related\n"
    "financial behavior and are important for\n"
    "credit risk analysis and model building."
)

ax2.text(0, 0.5, insight_text, fontsize=12, va='center')

plt.show()
```

#### Credit History Distribution by Loan Purpose

```python
# Chart - 9 visualization 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Clean column names 
bank_df.columns = bank_df.columns.str.strip().str.replace(' ', '_').str.lower()

# Define columns explicitly for your dataset 
loan_col = 'purpose_of_the_credit'
credit_col = 'credit_history'

#  Get top 10 loan types 
top_loan_types = bank_df[loan_col].value_counts().nlargest(10).index

# Filter dataset for top loan types 
filtered_bank_df = bank_df[bank_df[loan_col].isin(top_loan_types)]

#  Create pivot table for heatmap 
loan_credit_heatmap = filtered_bank_df.groupby([loan_col, credit_col]).size().unstack(fill_value=0)

#  Plot heatmap
fig, ax = plt.subplots(1, 1, figsize=(18, 10))

sns.heatmap(
    loan_credit_heatmap,
    annot=True,
    fmt='d',
    cmap='coolwarm',
    ax=ax,
    cbar_kws={'label': 'Number of Applicants'},
    linewidths=0.5
)

 # Styling function 
def apply_chart_styling(fig, ax, title='', subtitle='', insight_text=''):
    ax.set_title(f"{title}\n{subtitle}", fontsize=16, pad=20)
    ax.set_xlabel('Credit History', fontsize=12)
    ax.set_ylabel('Loan Purpose', fontsize=12)
    
    # Add insight text below plot
    plt.gcf().text(0.5, -0.05, insight_text, ha='center', fontsize=11, wrap=True)

# Apply styling and insight
apply_chart_styling(
    fig=fig,
    ax=ax,
    title='Credit History Distribution by Loan Purpose',
    subtitle='Analyzing Credit History Across the Top 10 Loan Purposes',
    insight_text='''
This chart shows the distribution of credit history for the top 10 loan purposes. 
Certain credit purposes, like "used car" or "education", may have more applicants with poor 
credit history, while loans for "new car" or "furniture" may show a more balanced distribution.
'''
)

plt.tight_layout()
plt.show()
```
